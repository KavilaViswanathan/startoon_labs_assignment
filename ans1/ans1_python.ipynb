```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data1 = np.loadtxt('Data_1.txt')
data2 = np.loadtxt('Data_2.txt')

def find_peaks(data):
    maxima = []
    minima = []
    for i in range(1, len(data) - 1):
        if data[i] > data[i - 1] and data[i] > data[i + 1]:
            maxima.append(i)
        elif data[i] < data[i - 1] and data[i] < data[i + 1]:
            minima.append(i)
    return maxima, minima

# Find peaks for Data_1
maxima_1, minima_1 = find_peaks(data1)

# Find peaks for Data_2
maxima_2, minima_2 = find_peaks(data2)

# Plot the results for Data_1
plt.figure(figsize=(10, 5))
plt.plot(data1, label='Data 1')
plt.scatter(maxima_1, data1[maxima_1], color='red', label='Maxima', marker='o')
plt.scatter(minima_1, data1[minima_1], color='blue', label='Minima', marker='o')
plt.title('Peaks in Data 1')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()

# Plot the results for Data_2
plt.figure(figsize=(10, 5))
plt.plot(data2, label='Data 2')
plt.scatter(maxima_2, data2[maxima_2], color='red', label='Maxima', marker='o')
plt.scatter(minima_2, data2[minima_2], color='blue', label='Minima', marker='o')
plt.title('Peaks in Data 2')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()

# Output the index numbers
print("Maxima indices for Data 1:", maxima_1)
print("Minima indices for Data 1:", minima_1)
print("Maxima indices for Data 2:", maxima_2)
print("Minima indices for Data 2:", minima_2)
```
